```python
import os
import pandas as pd

CSV_PATH = os.path.realpath(os.path.join(os.getcwd(), "data.csv")) #need to update csv
df = pd.read_csv(CSV_PATH).reset_index(drop=True).drop("id", axis=1)
df.head(20)
```

```python
for c in df.iloc[0: , :7]:
    for r in range(0, len(df)):
        df[c][r] = df[c][r].split("|")
        for i, x in enumerate(df[c][r]):
            if (df[c][r][i]) != "X":
                df[c][r][i] = int(df[c][r][i])
df.head()
```

```python
df["rxnTime"] = 0

#convert time since to start to time since beep for calibration sequence, computes and stores reaction time
for r in range(0, len(df)):
    dT = 2000
    tot = 0
    num = 9
    for i,t in enumerate(df.loc[r, "calib"]):
        if df.loc[r, "calib"][i] != "X":
            df.loc[r, "calib"][i] = abs(df.loc[r, "calib"][i] - dT)
            tot += df.loc[r, "calib"][i]
            dT = dT + 2000 - (i + 1) * 200
        else:
            num -= 1
    df.loc[r, "rxnTime"] = int(round(tot/num))

#convert time since to start to time since beep for rhytmic sequences
for id, c in enumerate(df.iloc[: , 2:7]):
    for r in range(0, len(df)):
        dT = 0
        for i, x in enumerate(df[c][r]):
            dT = dT + (60.0/float(df["freq"][r][id]))*1000
            if df[c][r][i] != "X":
                df[c][r][i] = abs(int(round(df[c][r][i] - dT)))
df.head()
```

```python
df[["aNum", "bNum", "cNum", "dNum", "eNum"]] = 0

#counts the number of beeps until time since beep is less than reaction time
for c in df.iloc[: , 2:7]:
    for r in range(0, len(df)):
        for i, x in enumerate(df[c][r]):
            if (x !="X" and x < df.loc[r, "rxnTime"]):
                df.loc[r, c+"Num"] = i + 1
                break;
df.head()
```

```python
freqDict = {
    40: [],
    50: [],
    60: [],
    70: [],
    80: [],
    90: [],
    100: []
}
freqDiffDict = {
    -10: [],
    -20: [],
    -30: [],
    -40: [],
    -50: [],
    -60: [],
    10: [],
    20: [],
    30: [],
    40: [],
    50: [],
    60: [],
}
idList = ["a", "b", "c", "d", "e"]

#collect beep counts into their respective frequencies
for r in range(0, len(df)):
    for id, f in enumerate(df.loc[r, "freq"]):
        freqDict[f].append(df.loc[r, idList[id]+"Num"])
print(freqDict)

#collects beep counts into a difference in frequency
for r in range(0, len(df)):
    for id, f in enumerate(df.loc[r, "freq"]):
        if id == 0:
            continue;
        freqDiffDict[df.loc[r, "freq"][id] - df.loc[r, "freq"][id - 1]].append(df.loc[r, idList[id]+"Num"])
print(freqDiffDict)
```
